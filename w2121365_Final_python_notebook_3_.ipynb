```python
# Bring in the tool to load and manage our dataset
import pandas as pd

# Load the classification data file into a table we can work with
data = pd.read_csv('Classification_Loan_Data.csv')
```

**Reused from: Code Reuse Session 3, Part 1 (Prompt 5, 6, 7)**

```python
# Set up the list of input features for the models
feature_cols = ['age', 'income', 'home_ownership', 'emplyment_length', 'loan_intent', 'loan_amount', 'loan_interest_rate', 'payment_default_on_file', 'credit_history_length']

# Assign input features to X
X = data[feature_cols] # Features

# Assign output variable to y
y = data['loan_approval_status']

# Bring in the tool to split the data into training and testing groups
from sklearn.model_selection import train_test_split

# Split the data 70/30, lock the random selection
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=42, stratify=y)
```

**Reused from: Code Reuse Session 2 (Prompts 9, 10, 11, 13, 17, 18, 19)**


```python
# Reused from Session 2 to evaluate Base Learner 1 (Logistic Regression) individually
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_lr = logreg.predict(X_test)

# Calculate and print the accuracy score
from sklearn.metrics import accuracy_score
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print ("The accuracy is: ", accuracy_lr)

from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
cm_lr = confusion_matrix(y_test, y_pred_lr, labels = logreg.classes_)
disp_lr = ConfusionMatrixDisplay(cm_lr, display_labels = logreg.classes_)
disp_lr.plot()

from sklearn.metrics import classification_report
print("Classification Report (Logistic Regression):\n", classification_report(y_test, y_pred_lr))

from sklearn.metrics import RocCurveDisplay
logreg_disp = RocCurveDisplay.from_estimator(logreg, X_test, y_test)
```

**Reused from: Code Reuse Session 2 (Prompts 27, 13, 17, 18, 19)**

*Note: The second base learner (Naive Bayes) is instantiated, trained, and evaluated individually on the test subset to establish its independent predictive performance before being combined into the final ensemble.*

```python
# Reused from Session 2 to evaluate Base Learner 2 (Naive Bayes) individually
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)

# Calculate and print the accuracy score
from sklearn.metrics import accuracy_score
accuracy_nb = accuracy_score(y_test, y_pred_nb)
print ("The accuracy is: ", accuracy_nb)

from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
cm_nb = confusion_matrix(y_test, y_pred_nb, labels = nb.classes_)
disp_nb = ConfusionMatrixDisplay(cm_nb, display_labels = nb.classes_)
disp_nb.plot()

from sklearn.metrics import classification_report
print("Classification Report (Naive Bayes):\n", classification_report(y_test, y_pred_nb))

from sklearn.metrics import RocCurveDisplay
nb_disp = RocCurveDisplay.from_estimator(nb, X_test, y_test)
```

**Reused from: Code Reuse Session 3, Part 1 (Prompts 8, 9, 10, 11) & Code Reuse Session 2 (Prompts 9 & 27)**

```python
# (Reused from Session 2) Bring in the Logistic Regression model
from sklearn.linear_model import LogisticRegression

# (Reused from Session 2) Bring in the Naive Bayes model
from sklearn.naive_bayes import GaussianNB

# Bring in the tool to combine multiple models (Ensemble)
from sklearn.ensemble import VotingClassifier

# Set up the list of models we want to combine
base_learners=[('LR', LogisticRegression()), ('NB', GaussianNB())]

# Instantiate the voting ensemble algorithm with soft voting
ensemble_learner = VotingClassifier(base_learners, voting='soft')

# Train the combined model using the training data
ensemble_learner.fit(X_train, y_train)

# Make predictions on isolated data (test subset)
y_pred_ensemble_learner = ensemble_learner.predict(X_test)
```

**Reused from: Code Reuse Session 3, Part 1 (Prompts 12, 13, 14, 15)**

```python
# Bring in the tool to check model accuracy
from sklearn.metrics import accuracy_score

# Calculate the accuracy on the test data
ensemble_learner_accuracy = accuracy_score(y_test, y_pred_ensemble_learner)

# Print the calculated accuracy score
print ("The voting ensemble classifier accuracy is: ", ensemble_learner_accuracy)

# Bring in the tools to make a confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Calculate the confusion matrix for the ensemble model
ensemble_learner_cm_test = confusion_matrix(y_test, y_pred_ensemble_learner, labels=ensemble_learner.classes_)

# Set up the visual chart for the matrix
ensemble_learner_disp = ConfusionMatrixDisplay(ensemble_learner_cm_test, display_labels=ensemble_learner.classes_)

# Show the confusion matrix
ensemble_learner_disp.plot()

# Add a clear title to the chart
ensemble_learner_disp.ax_.set_title("Ensemble Learner Confusion Matrix")

# Bring in the tool for the detailed classification report
from sklearn.metrics import classification_report

# Print the report to check Precision, Recall, and F1-scores
print("Ensemble Learner Classification Report \n", classification_report(y_test, y_pred_ensemble_learner))

# Bring in the tool to plot the ROC curve
from sklearn.metrics import RocCurveDisplay

ensemble_learner_roc = RocCurveDisplay.from_estimator(ensemble_learner, X_test, y_test)
```

**Reused from: Code Reuse Session 3, Part 2 (Prompts 1, 2, 3, 4, 5)**

```python
# Maximum Loan Amount Regression Modelling

# Bring in the tool to load and manage our dataset
import pandas as pd

# Load the regression data file into a table we can work with
dataset = pd.read_csv('Regression_Loan_Data.csv')

# Show the structure and data types for all columns in the dataset
dataset.info()

# Separate the input features by dropping the target and status columns
X = dataset.drop(['max_allowed_loan', 'loan_approval_status'], axis=1)

# Separate the specific target we are trying to predict (Maximum Loan Amount)
y = dataset['max_allowed_loan']

# Bring in the tool to split the data into training and testing groups
from sklearn.model_selection import train_test_split

# Split the data 70/30 and lock the random selection to ensure consistent testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Reused from: Code Reuse Session 3, Part 2 (Prompts 6, 7, 8, 9)**


```python
# Bring in the Decision Tree Regressor model
from sklearn.tree import DecisionTreeRegressor

# Set up the fully grown Decision Tree model (DT-1)
DT_regressor = DecisionTreeRegressor()

# Train the model using the training data
DT_regressor.fit(X_train, y_train)

# Make predictions on your isolated dataset
y_pred = DT_regressor.predict(X_test)

# Bring in the tools to check the model's error and accuracy
from sklearn import metrics

# Calculate and present the Mean Absolute Error
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))

# Calculate and present the Mean Squared Error
print('MSE:', metrics.mean_squared_error(y_test, y_pred))

# Calculate and present the R-squared score
print('R2:', metrics.r2_score(y_test, y_pred))
```

**Reused from: Code Reuse Session 3, Part 2 (Prompts 10, 11, 12)**

```python
# Bring in the tools to draw and display the decision tree
from sklearn import tree
from matplotlib import pyplot as plt

# Initiate the figure size for the plot
Tree_figure = plt.figure(figsize=(200,200))

# Plot your decision tree for the maximum loan amount regression model
DT_Graph = tree.plot_tree(DT_regressor, feature_names=list(X_train.columns), filled=True)

# Save your DT graph as a high-resolution image
Tree_figure.savefig("decision_tree_fully_grown.svg")
```

**Reused from: Code Reuse Session 3, Part 2 (Prompt 13, 9, 11, 12)**

```python
# Prune the regression DT to 4 levels
DT_regressor_pruned = DecisionTreeRegressor(max_depth=4)

# Train the pruned model
DT_regressor_pruned.fit(X_train, y_train)

# Perform predictions on your isolated dataset
y_pred_pruned = DT_regressor_pruned.predict(X_test)

# Calculate and present the Mean Absolute Error for the pruned tree
print('MAE:', metrics.mean_absolute_error(y_test, y_pred_pruned))

# Calculate and present the Mean Squared Error for the pruned tree
print('MSE:', metrics.mean_squared_error(y_test, y_pred_pruned))

# Calculate and present the R-squared score for the pruned tree
print('R2:', metrics.r2_score(y_test, y_pred_pruned))

# Initiate the figure size for the pruned tree plot
Tree_figure_pruned = plt.figure(figsize=(50,50))

# Plot your pruned decision tree
DT_Graph_pruned = tree.plot_tree(DT_regressor_pruned, feature_names=list(X_train.columns), filled=True)

# Save your pruned DT graph as a high-resolution image
Tree_figure_pruned.savefig("decision_tree_pruned.svg")
```
